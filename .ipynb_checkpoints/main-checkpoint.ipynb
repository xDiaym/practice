# %matplotlib inline
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

```python
df = pd.read_json("./dataset.json")
df.head()
```

```python
mask = df.columns.str.contains("temp") & ~df.columns.str.contains("CPU|mediana|tempCT")
df.loc[:,mask].mean()
```

```python
data = df.dropna(subset="BMP280_temp")
data["date"] = pd.to_datetime(data["date"], format="%Y-%m-%d %H:%M:%S")
dmin, dmax = data["date"].agg(["min", "max"])

days = mdates.drange(dmin, dmax, dt.timedelta(days=7))

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(data["date"], data["BMP280_temp"])


data = df.dropna(subset="BMP280_0_temp")
data["date"] = pd.to_datetime(data["date"], format="%Y-%m-%d %H:%M:%S")
ax.plot(data["date"][::8], data["BMP280_0_temp"][::8], linewidth=1)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%d%b %H:%M'))
ax.xaxis.set_major_locator(mdates.HourLocator(interval=12))
ax.tick_params(axis='x', labelrotation=70)
fig.legend(["BMP280", "BMP280_0"])

plt.show()
```
